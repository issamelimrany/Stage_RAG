# Hybrid Search

## BM25 Retriever - Sparse retriever

```python
!pip install -U langchain-huggingface
```

```python
from langchain.retrievers import BM25Retriever, EnsembleRetriever
from langchain.schema import Document

from langchain.vectorstores import Chroma
from langchain.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings
embedding=HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

```

```python
# Example document list
doc_list = [
    "I enjoy hiking in the mountains",
    "Mountain biking is an exciting sport",
    "Hiking and biking are popular outdoor activities",
    "I have a collection of hiking gear",
    "Adventure sports often include activities like hiking and biking"
]
```

```python
# initialize the bm25 retriever and faiss retriever
bm25_retriever = BM25Retriever.from_texts(doc_list)
bm25_retriever.k = 2
```

```python
bm25_retriever.get_relevant_documents("hiking")
```

```python
bm25_retriever.get_relevant_documents("activity")
```

```python
bm25_retriever.dict
```

## Embeddings - Dense retrievers FAISS

```python
faiss_vectorstore = FAISS.from_texts(doc_list, embedding)
faiss_retriever = faiss_vectorstore.as_retriever(search_kwargs={"k": 2})
```

```python
faiss_retriever.get_relevant_documents("activity")
```

## Ensemble Retriever

```python
# initialize the ensemble retriever
ensemble_retriever = EnsembleRetriever(retrievers=[bm25_retriever, faiss_retriever],
                                       weights=[0.5, 0.5])
```

```python
docs = ensemble_retriever.get_relevant_documents("activity")
docs
```

```python
docs = ensemble_retriever.get_relevant_documents("sport")
docs
```
